# Analise dos dados por meio de medias



**MEDIA DE ESCANTEIOS**

```python
time=df['team_name']
media=(df['corners_total']/df['matches_played']).round(2)

plt.figure(figsize=(15, 5))
bars=plt.bar(time, media)

plt.title('Media de escanteios por jogo')
plt.xlabel('Time')
plt.ylabel('Escanteios')

# Adicionando os valores sobre as barras de forma simples
plt.bar_label(bars)

plt.grid(True, axis='y')
plt.xticks(rotation=90)  # Para girar os nomes dos times
plt.show()



print(f'A media de escanteios por partida durante o campeonato foi de {(media.sum()/df['team_name'].count()).round(1)}')

lista = df['corners_total'].tolist()
```

**CHUTES AO GOL**

```python
time=df['team_name']
media=(df['shots_on_target']/df['matches_played']).round(2)

plt.figure(figsize=(15, 5))
bars=plt.bar(time, media)

plt.title('Media de chutes ao gol por jogo')
plt.xlabel('Time')
plt.ylabel('Chutes ao gol')

# Adicionando os valores sobre as barras de forma simples
plt.bar_label(bars)

plt.grid(True, axis='y')
plt.xticks(rotation=90)  # Para girar os nomes dos times
plt.show()


print(f'A media de chutes ao gol por partida durante o campeonato foi de {(media.sum()/df['team_name'].count()).round(1)}')
```

**FALTAS COMETIDAS**

**CARTOES**
